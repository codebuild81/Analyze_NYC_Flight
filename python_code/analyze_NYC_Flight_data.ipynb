```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Read .csv file data using pandas dataframe

```python
#Read data from .csv file
flightdatacsv = pd.read_csv(your_local_path+'flight_data.csv')

#print record count
print (flightdatacsv.iloc[:].count())

#see some records using head()
flightdatacsv.head()
```

```python
flightdatacsv.describe()
```

```python
#Q1. Which flight is having maximum departure delay?

max_dep_delay = flightdatacsv.iloc[:,5].max()
idx = flightdatacsv[flightdatacsv['dep_delay'] == max_dep_delay]
print (idx[['dep_delay','flight', 'carrier']])
```

```python
#Q2. Find out delay departure count, on time departure count, early departure count 

#delay departure count data
flight_dep_delay = flightdatacsv[(flightdatacsv['dep_delay'] > 0)]
print ('Delay departure count: ',flight_dep_delay['dep_delay'].count())

#on time departure count data
flight_on_time = flightdatacsv[(flightdatacsv['dep_delay'] == 0)]
print ('On time departure count: ',flight_on_time['dep_delay'].count())

#early departure count
flight_early_dep = flightdatacsv[(flightdatacsv['dep_delay'] < 0)]
print ('Early departure count: ',flight_early_dep['dep_delay'].count())
flight_early_dep
```

```python
#Q3. Which flight is having maximum number of departure delay count?

flightwise_dep_delay_count = flight_dep_delay.groupby(['flight']).count()
print(flightwise_dep_delay_count.shape)

m = flightwise_dep_delay_count['dep_delay'].max()
m
flight_dep_delay_max_count = flightwise_dep_delay_count[flightwise_dep_delay_count['dep_delay'] == m]
flightwise_dep_delay_count
print (flight_dep_delay_max_count.iloc[:,5].reset_index(name = 'delay_count'))
```

```python
#Q4. Plot Flight vs. departure delay count
#Hold data set after reset of index
hold_dataset = flightwise_dep_delay_count.reset_index()

#Considering 50 flight record set for most delay count
plt_dataset = hold_dataset[['flight','dep_delay']].sort_values(['dep_delay'], ascending = False).iloc[:50]
#print (plt_dataset)

# Plot Bar chart   
plt_dataset.plot(x='flight', y='dep_delay', kind='bar', figsize=(20, 20)) 
plt.title("Flight Vs. Departure delay count")
plt.xlabel("Flight")
plt.ylabel("Departure delay count")
plt.show()
```

```python
#Q5. Total flight departure count for each origin port.

total_flight_dep_originwise = flightdatacsv.groupby(['origin']).size()

print(total_flight_dep_originwise)
print('Total flight departure count for each origin port.')
tot_count_each_origin = total_flight_dep_originwise.reset_index(name = 'count')
tot_count_each_origin
```

```python
#Q6. Find total on time departure count in each origin port
flight_on_time.head()
total_ontime_flight_dep_originwise = flight_on_time.groupby(['origin']).size()
print(total_ontime_flight_dep_originwise)

print('Total On time flight departure count for each origin port.')
tot_ontime_count_each_origin = total_ontime_flight_dep_originwise.reset_index(name = 'count')
tot_ontime_count_each_origin

```

```python
#Merge two data set into one based on common column
mdata = pd.merge(tot_count_each_origin,tot_ontime_count_each_origin,on='origin',how='inner')
mdata
```

```python
#Q7. Calculate percentage for each origin port

mdata['Percentage'] = mdata.iloc[:,2]/mdata.iloc[:,1]*100
mdata
```

```python
#Q8. Find best airports in terms of time departure percentage (%).
mdata[mdata.iloc[:,3] == mdata.iloc[:,3].max()]
```

```python
#Q9. Aircraft speed analysis.
flightdatacsv[['distance']].count()
flightdatacsv[['hour']].count()

#speed of each flight in Miles
flightdatacsv['speed'] = flightdatacsv.loc[:,'distance']/(flightdatacsv.loc[:,'hour']*60 + flightdatacsv.loc[:,'minute'])*60
flightdatacsv.head()
```

```python
#Q10. Find Maximum speed of the flight & which flight
flightdatacsv[flightdatacsv.loc[:,'speed'] == flightdatacsv.loc[:,'speed'].max()]
```

```python
#Q11. On time arrival percentage(%) analysis.
flightdatacsv.head()
```

```python
#delay arrival count data

per_flt_delay_arr_cnt = flightdatacsv[(flightdatacsv.loc[:,'arr_delay'] > 0)].groupby(['flight']).size()
per_flt_delay_arr_cnt_1 = per_flt_delay_arr_cnt.reset_index(name = 'delay_arr_count')
per_flt_delay_arr_cnt_1

#on time arrival count data
per_flt_ontime_arr_cnt = flightdatacsv[(flightdatacsv.loc[:,'arr_delay'] == 0)].groupby(['flight']).size()
per_flt_ontime_arr_cnt_1 = per_flt_ontime_arr_cnt.reset_index(name = 'ontime_arr_count')
per_flt_ontime_arr_cnt_1

#early arrival count
per_flt_early_arr_cnt = flightdatacsv[(flightdatacsv.loc[:,'arr_delay'] < 0)].groupby(['flight']).size()
per_flt_early_arr_cnt_1 = per_flt_early_arr_cnt.reset_index(name = 'early_arr_count')
per_flt_early_arr_cnt_1

#Merge three data set into one based on common column flight
mdata_three = pd.merge(pd.merge(per_flt_delay_arr_cnt_1,per_flt_ontime_arr_cnt_1,on='flight',how='outer'),per_flt_early_arr_cnt_1,on='flight',how='outer')
mdata_three
mdata_three = mdata_three.fillna(0)
mdata_three

#On time arrival percentage(%)
mdata_three['ontime_arr_percent'] = (mdata_three.loc[:,'ontime_arr_count']/(mdata_three.loc[:,'delay_arr_count'] + mdata_three.loc[:,'ontime_arr_count'] + mdata_three.loc[:,'early_arr_count']))*100
mdata_three
pd.set_option('precision',2)
mdata_three
```

```python
#Q12. Plot Flight Vs. Flight On Time
#Considering 50 flight record set for best on time percentage.
plt_ontime_arr_percent = mdata_three[['flight','ontime_arr_percent']].sort_values(['ontime_arr_percent'], ascending = False).iloc[:50]
#print (plt_dataset)

# Plot Bar chart   
plt_ontime_arr_percent.plot(x='flight', y='ontime_arr_percent', kind='bar', figsize=(20, 20)) 
plt.title("Flight Vs. Flight On Time %")
plt.xlabel("Flight")
plt.ylabel("Flight On Time ")
plt.show()
plt_ontime_arr_percent
```
